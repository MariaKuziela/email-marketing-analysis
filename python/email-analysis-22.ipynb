```python
'''success rate według typu kampanii (wniosek nr 1 z SQL)

Potrzebujesz:

Przefiltrować df — wykluczyć Email_Status = 2
Pogrupować po Email_Campaign_Type
Policzyć success rate dla każdej grupy
Narysować wykres słupkowy

'''

import pandas as pd
import matplotlib.pyplot as plt
#import numpy as np


df = pd.read_csv('/Users/mariamagdalena/Desktop/Data Engineering/projects-sql-python/email-campaign-sme/Train_psolI3n.csv')
df = df[df['Email_Status'] != 2]
df_grouped = df.groupby('Email_Campaign_Type').agg(
    Success_Rate=('Email_Status', 'mean'),
    Total_Emails=('Email_Status', 'count')
).reset_index()

#print(df_grouped)

x = df_grouped['Email_Campaign_Type']
y = df_grouped['Success_Rate']

plt.title('Success Rate by Email Campaign Type')
plt.xlabel('Email Campaign Type')
plt.ylabel('Success Rate')

plt.bar(x, y)
plt.show()
```
