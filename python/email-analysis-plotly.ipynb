```python
import plotly.express as px
import pandas as pd

df = pd.read_csv('/Users/mariamagdalena/Desktop/Data Engineering/projects-sql-python/email-campaign-sme-dataset/Train_psolI3n.csv')
df = df[df['Email_Status'] != 2]

#grupowanie po typie kampanii
df_grouped = df.groupby('Email_Campaign_Type').agg(
    Success_Rate=('Email_Status', 'mean'),
    Total_Emails=('Email_Status', 'count')
).reset_index()

#print(df_grouped)

#tworzenie wykresu
x = df_grouped['Email_Campaign_Type']
y = df_grouped['Success_Rate']

fig = px.bar(df_grouped, x='Email_Campaign_Type', y='Success_Rate', title='Success Rate by Email Campaign Type',
                 color='Success_Rate',
                color_continuous_scale='ylgnbu',
                hover_data=['Total_Emails'],)


fig.update_layout(
    xaxis_title='Typ kampanii',
    yaxis_title='Success Rate',
    xaxis=dict(tickmode='array', tickvals=[1, 2, 3]),
    plot_bgcolor='#F9F9F9',
    yaxis=dict(tickformat='.0%')
)


fig.show()

```

Note: Campaign Type 1 has a significantly smaller sample size (563 emails vs 47k+) and should be interpreted with caution.

```python
'''
success rate według długości emaila (Word_Count)


'''


import plotly.express as px
import pandas as pd



df = pd.read_csv('/Users/mariamagdalena/Desktop/Data Engineering/projects-sql-python/email-campaign-sme-dataset/Train_psolI3n.csv')
df = df[df['Email_Status'] != 2]

print(df['Word_Count'].describe())

#przedziały dla Word Count
df['Word_Count_Group'] = pd.cut(df['Word_Count'], bins=[40, 525, 694, 881, 1317], labels=['40-524', '525-693', '694-880', '881-1316'])

#grupowanie po Word Count
df_word_grouped = df.groupby('Word_Count_Group', observed=True).agg(
    Success_Rate=('Email_Status', 'mean'),
    Total_Emails=('Email_Status', 'count')
).reset_index()

print(df['Word_Count_Group'].value_counts(dropna=False))
print(df_word_grouped)

#tworzenie wykresu
x = df_word_grouped ['Word_Count_Group']
y = df_word_grouped ['Success_Rate']

fig = px.bar(df_word_grouped, x='Word_Count_Group', y='Success_Rate', title='Success Rate by Email Word Count Group',
                category_orders={'Word_Count_Group': ['40-524', '525-693', '694-880', '881-1316']},
                color='Success_Rate',
                color_continuous_scale='greens',
                hover_data=['Total_Emails'],)


fig.update_layout(
    xaxis_title='Word Count Group',
    yaxis_title='Success Rate',
    plot_bgcolor='white',
    yaxis=dict(tickformat='.0%')
)


fig.show()

```

```python
'''--- Która lokalizacja klientów (Customer_Location) ma najwyższy success rate?
grupowanie po lokalizacji, liczenie liczby emaili, liczby pozytywnych reakcji i obliczanie wskaźnika (procent) sukcesu


'''


import plotly.express as px
import pandas as pd



df = pd.read_csv('/Users/mariamagdalena/Desktop/Data Engineering/projects-sql-python/email-campaign-sme-dataset/Train_psolI3n.csv')
df = df[df['Email_Status'] != 2]

print(df['Customer_Location'].describe())


df_c_location_grouped = df.groupby('Customer_Location', observed=True).agg(
    Success_Rate=('Email_Status', 'mean'),
    Total_Emails=('Email_Status', 'count')
).reset_index()

print(df['Customer_Location'].value_counts(dropna=False))
print(df_c_location_grouped)


x = df_c_location_grouped ['Customer_Location']
y = df_c_location_grouped ['Success_Rate']

fig = px.bar(df_c_location_grouped, x='Customer_Location', y='Success_Rate', title='Success Rate by Customer Location',
                color='Success_Rate',
                color_continuous_scale='burgyl',
                hover_data=['Total_Emails'],)


fig.update_layout(
    xaxis_title='Customer Location',
    yaxis_title='Success Rate',
    plot_bgcolor='white',
    yaxis=dict(tickformat='.0%')
)


fig.show()

```

```python
'''
success rate według liczby linków w emailu (Link_Count)



'''


import plotly.express as px
import pandas as pd



df = pd.read_csv('/Users/mariamagdalena/Desktop/Data Engineering/projects-sql-python/email-campaign-sme-dataset/Train_psolI3n.csv')
df = df[df['Email_Status'] != 2]



#przedziały dla liczby linków
df['Link_Count_Group'] = pd.cut(df['Total_Links'], bins=[0, 5, 15, 16], labels=['1-5_LINKS', '6-15_LINKS', '>16_LINKS'])

#grupowanie po Link Count
df_link_grouped = df.groupby('Link_Count_Group', observed=True).agg(
    Success_Rate=('Email_Status', 'mean'),
    Total_Emails=('Email_Status', 'count')
).reset_index()

print(df['Link_Count_Group'].value_counts(dropna=False))
print(df_link_grouped)

#tworzenie wykresu
x = df_link_grouped ['Link_Count_Group']
y = df_link_grouped ['Success_Rate']

fig = px.bar(df_link_grouped, x='Link_Count_Group', y='Success_Rate', title='Success Rate by Email Link Count Group',
                category_orders={'Link_Count_Group': ['1-5_LINKS', '6-15_LINKS', '>16_LINKS']},
                color='Success_Rate',
                color_continuous_scale='magenta',
                hover_data=['Total_Emails'],)


fig.update_layout(
    xaxis_title='Link Count Group',
    yaxis_title='Success Rate',
    plot_bgcolor='light grey',
    yaxis=dict(tickformat='.0%')
)


fig.show()

```
